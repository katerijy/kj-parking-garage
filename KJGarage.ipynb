```python
class KJGarage():
    
    def __init__(self, current_ticket, parking_space = 10, ticket_count = 10):
        self.parking_space = parking_space #list of parking spaces 1-10
        self.current_ticket = current_ticket #(true or false)
        self.current_ticket = {1:["free", "false"], 2:["free", "false"], 3:["free", "false"], 4:["free", "false"], 
                               5:["free", "false"], 6:["free", "false"], 7:["free", "false"], 8:["free", "false"], 
                               9:["free", "false"], 10:["free", "false"]}
        self.ticket_count = ticket_count
        
    def home(self):
        self.options = input("What would you like to do: Take a ticket, Pay, or Leave? ")
        if self.options.lower() == "take a ticket" or self.options.lower() == "take ticket" or self.options.lower() == "ticket":
            self.takeTicket()
        elif self.options.lower() == "pay":
            self.payForParking()
        elif self.options.lower() == "leave":
            self.leaveGarage()
        else:
            print("Error - Not a valid response")
            self.home()
        
    def enterGarage(self):
        print(f"Welcome to KJ Parking Garage!")
        self.home()
        
    def takeTicket(self):
        if self.ticket_count == 0:
            print("Sorry, we're full!")
            self.home()
        else:
            for key in self.current_ticket:
                if self.current_ticket[key][0] == "free":
                    self.current_ticket[key][0] = "taken"
                    print(f'You have ticket #{key}')
                    break
            self.parking_space -= 1
            self.ticket_count -= 1  
        print("Here is your ticket. Park safely")
        self.home()
      
    def payForParking(self):
        question = int(input("What is your ticket number? ")) 
        if self.current_ticket[question][0] != "taken":
            print("Error - This ticket is not in use")
        else:
            if question > 10:
                print("Error - Not a valid response")
                question
            else:
                pay_ticket1 = input("You owe $10. You can choose to add $1 as a gift to the city homeless shelter. \n Please enter the amount you are paying here: $") 
                while True:
                    if int(pay_ticket1) == 10:
                        self.ticket_count += 1
                        self.current_ticket[question][1] = "true"
                        print("Your ticket has been paid and you have 15 minutes to exit the garage.")
                        break
                    if int(pay_ticket1) > 10:
                        self.ticket_count += 1
                        self.current_ticket[question][1] = "true"
                        print("Thank you for your donation!")
                        print("Your ticket has been paid and you have 15 minutes to exit the garage.")
                        break
                    elif int(pay_ticket1)<10:
                        print("Error - Please pay at least $10 to exit the garage.")
                        break
                    pay_ticket1
        self.home()
  
    def leaveGarage(self):
        question2 = int(input("What is your ticket number? "))
        if question2 > 10:
            print("Error - Not a valid response")
            question2
        else:
            if self.current_ticket[question2][0] == "taken" and self.current_ticket[question2][1] == "true":
                print("Your visit is complete. Thank You, have a nice day")
                self.current_ticket[question2][0] == "free" 
                self.current_ticket[question2][1] == "false"
            elif self.current_ticket[question2][0] == "taken" and self.current_ticket[question2][1] == "false":
                self.payForParking()
            else:
                print("Error - This ticket is not in use")
        self.home()

                    
customer1 = KJGarage(1)
customer1.enterGarage()

                
```
